### Prepare data

```python
#Currently only 2 files
bfiles = ["bkgd6", "sig3"]
njets = 2
features = ['Split12', 'Split23', 'D2', 'C2', 'tau32', 'pt']
jet1 = ['jet1_Split12', 'jet1_Split23', 'jet1_D2', 'jet1_C2', 'jet1_tau32', 'jet1_pt']
jet2 = ['jet2_Split12', 'jet2_Split23', 'jet2_D2', 'jet2_C2', 'jet2_tau32', 'jet2_pt']
data = dict.fromkeys(['Split12', 'Split23', 'D2', 'C2', 'tau32', 'pt', 'label'])
first = True

#For Each Jet 1 and 2, seperate and put in data
for x in range(njets):
    curjet = "jet" + str(x + 1)
    for filename in tqdm(bfiles):
        tmp = (globals()[filename])['PostSel'].arrays((globals()[curjet]), library="np")
        for f in range(len(features)):
            #rename all features to generalized name, from jet1 -> jet2
            tmp[features[f]] = tmp.pop((globals()[curjet])[f])
        #add column to label each particle (0 -> background, 1 -> signal)
        lb_tmp = 0 if "bkgd" in filename else 1
        tmp['label'] = np.full(len(tmp[features[0]]), lb_tmp)
        #merge current dictionary with data (unless first pass)
        if not first:
            values = zip(tmp.values(), data.values())
            data = dict(zip(keys, values))
            for lbl in data.keys():
                data[lbl] = np.concatenate((data[lbl][0], data[lbl][1]))
        else:
            data = tmp
        first = False

df = pd.DataFrame(data)

X = df.loc[:, features]
y = df.label
```
